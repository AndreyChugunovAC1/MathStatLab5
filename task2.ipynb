$ \mu_i - \text{среднее влияние фактора на уровне i} $

$ H_0: \mu_1 = \mu_2 = ... = \mu_I$

$ H_1: \overline H_0$

$S_\beta^2 = \sum_i J_i (\overline{Y_{i*}} - \overline{Y})^2$

$\overline{Y_{i*}} = \frac 1 {J_i} \sum^{J_i}_i y_{ij}$

$S_w^2 = \sum_{ij} (\overline{Y_{ij}} - \overline{Y_{i*}})^2$

$F = \frac{n-I}{I-1} \frac{S_B^2}{S_W^2} \sim F(n-I, I-1)$

```python
import numpy as np
import pandas as pd
from scipy.stats import f as f_dist

df = pd.read_csv("/content/exams_dataset.csv")

df["sum_scores"] = df["math score"] + df["reading score"] + df["writing score"]

y_col = "sum_scores"
g_col = "race/ethnicity"

y = df[y_col].to_numpy()
y_bar = np.nanmean(y)

grp = df.groupby(g_col)[y_col]
means = grp.mean()
counts = grp.size()
I = means.shape[0]
n = int(counts.sum())

S_B = float((counts * (means - y_bar) ** 2).sum())


S_W = float(grp.apply(lambda s: ((s - s.mean()) ** 2).sum()).sum())

df_between = I - 1
df_within = n - I

MSB = S_B / df_between
MSW = S_W / df_within

p_value = 1 - f_dist.cdf(MSB / MSW, df_between, df_within)


print(f"S_B = {S_B:.6f}, S_W = {S_W:.6f}")
print(f"df_between = {df_between}, df_within = {df_within}")
print(f"M_B = {MSB:.6f}, M_W = {MSW:.6f}")
print(f"p-value = {p_value:.6g}")
```
